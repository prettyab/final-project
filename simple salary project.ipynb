```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/DELL/Desktop/whatsapp/Salary Data.csv")
data
```

```python
df=data[['Years of Experience','Salary']]
df
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(inplace=True,drop=True)
```

```python
df.head()
```

```python
df=df.fillna({
    'Years of Experience':df['Years of Experience'].mean(),
    'Salary':df['Salary'].mean()
})
df
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['Salary'])
y=df['Salary']
X
```

```python
y
```

```python
X=df['Years of Experience'].values.reshape(-1,1)
y=df['Salary'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
next_salary=model.predict([[0.7]])
print(int(next_salary))
```

```python
model.score(X_train,y_train)
```
